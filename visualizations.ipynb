```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
log = pd.read_csv("log.txt", header = None, names = ["iteration", "train loss", "val loss", "hella swag", "train loss2", "val loss2", "hella swag2"])
```

```python
log
```

```python
train_log = []
val_log = []
hls_log = []

iteration_counter = 0
for index, row in log.iterrows():
    train_log.append((float(row["train loss"].split(":")[1]), iteration_counter))
    if not(row["val loss"] == "val_loss:None"):
        val_log.append((float(row["val loss"].split(":")[1]), iteration_counter))
    if not(row["hella swag"] == "HellaSwag_acc:None"):
        hls_log.append((float(row["hella swag"].split(":")[1]), iteration_counter))
    iteration_counter +=1
    if not(pd.isnull(row["train loss2"])):
        train_log.append((float(row["train loss2"].split(":")[1]), iteration_counter))
        iteration_counter +=1
```

```python
#See HLS final perfromance in other notebook...
```

```python
train_x = [pair[1] for pair in train_log]
train_y = [pair[0] for pair in train_log]

val_x = [pair[1] for pair in val_log]
val_y = [pair[0] for pair in val_log]

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot((train_x), (train_y), label='Training Loss', color='blue')
plt.plot((val_x), (val_y), label='Validation Loss', color='red')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Loss over Iterations')
plt.legend()
plt.grid(True)
plt.show()
```
